```julia
using PyPlot
using Statistics


using PyCall
np = pyimport("numpy")
include("Ising 2D.jl") #Contiene las funciones del modelo de Ising
```

```julia
#con 1millon jala en 3 seegundos
@time begin
L=100
T=1
num_pasos = 100000
ener, mag, m = simulacion_montecarlo(L::Int, T, num_pasos::Int,"caliente")
end
```

```julia
figure(figsize=(15,6))

subplot(1,2,1)
#plot(enerF,label ="fria")
plot(ener/50^2,label="caliente")
title("Energia")
legend()

subplot(1,2,2)
#plot(magF,label="fria")
plot(mag/50^2,label = "caliente")
title("Magnetización")
legend()
```
