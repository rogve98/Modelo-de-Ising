```julia
using Plots
using Statistics
using LaTeXStrings

using PyCall
np = pyimport("numpy")
include("Ising general.jl") #Contiene las funciones del modelo de Ising
```

```julia
include("Ising general.jl") #Contiene las funciones del modelo de Ising
microEstados50 = Configuraciones(50,0.5)
microEstados100 = Configuraciones(100,0.5)
microEstados200 = Configuraciones(200,0.5)
```

```julia
@time begin
    T = (0.1:0.05:2.5) 
    r50 = cantidadesObservables(microEstados50,T,200000,"fria")
end
```

```julia
@time begin
    T = (0.1:0.05:2.5) 
    r100 = cantidadesObservables(microEstados100,T,200000,"fria")
end
```

```julia
@time begin
    T = (0.1:0.05:2.5) 
    r200 = cantidadesObservables(microEstados200,T,200000,"fria")
end
```

```julia
plot(T,r50[2],w=2, color = "orange", label = L"Red de $N=50$", 
    title = "Magnetización",
    xlabel = "Temperatura [T]",
    ylabel = L"M")
plot!(T,r100[2],w=2, color = "magenta", label = L"Red de $N=100$")
plot!(T,r200[2],w=2, color = "violet", label = L"Red de $N=200$")

scatter!(T,r50[2],w=2, color = "orange", label = "")
scatter!(T,r100[2],w=2, color = "magenta", label = "")
scatter!(T,r200[2],w=2, color = "violet", label = "")

vline!([1.134592657106511],w=2, color = "red", label = L"T_C\approx 1.1346",s=:dash)

#savefig("susMagn")
```
